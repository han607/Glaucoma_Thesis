```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# pd.reset_option('display.max_rows')

# Importing the dataset
# two_intervals = pd.read_csv('dataset_interval_two.csv')
three_intervals = pd.read_csv('dataset_interval_three.csv')
four_intervals = pd.read_csv('dataset_interval_four.csv')
```

## Three Visits

### MD analysis

```python
from sklearn.linear_model import LinearRegression

difference_MD_3 = []
Third_y_predictions = []
Third_y_actuals = []
for i in range(0, len(three_intervals)-2, 3):
    First_X = three_intervals['days_from_baseline'][i]
#     print(First_X)
    First_Y = three_intervals['md'][i]
#     print(First_Y)
    Second_X = three_intervals['days_from_baseline'][i+1]
#     print(Second_X)
    Second_Y = three_intervals['md'][i+1]
#     print(Second_Y)
    Third_X = three_intervals['days_from_baseline'][i+2]
#     print(Third_X)
    Third_Y = three_intervals['md'][i+2]
#     print(Third_Y)
    
    
    model = LinearRegression()
    model.fit([[First_X], [Second_X]],[[First_Y], [Second_Y]])
    Third_y_pred = model.predict(Third_X)
    Third_y_actuals.append(Third_Y)
    Third_y_predictions.append(Third_y_pred)
    
#     differences = abs((Third_y_pred - Third_Y)/Third_Y)*100
    differences = abs((Third_y_pred - Third_Y))

    difference_MD_3.append([First_Y, Second_Y, Third_Y, Third_y_pred.item(0), differences.item(0)])

difference_MD_3 = pd.DataFrame(difference_MD_3)
difference_MD_3.columns = ['Visit_1_MD', 'Visit_2_MD', 'Visit_3_MD_Actual', 'Visit_3_MD_Predicted', 'Absolute difference']
difference_MD_3.describe()
```

```python
Third_y_predictions = np.array(Third_y_predictions)
Third_y_predictions = Third_y_predictions.ravel()
np.savetxt('Y_predictions_Linear_3', Third_y_predictions)
Third_y_actuals = np.array(Third_y_actuals)
Third_y_actuals
np.savetxt('Y_actuals_Linear_3', Third_y_actuals)
```
